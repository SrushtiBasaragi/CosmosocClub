# **Comparing performance with lists**

```python
size = 10000
```

```python
%%time
ls = list(range(size))
for i in range(size):
    ls[i] = ls[i] * ls[i]
```

```python
%%time
ls_ = list(range(size))
ls_ = [item * item for item in ls_]
```

```python
%%time
list_ = list(range(size))
list_ = map(lambda x: x * x, list_)
```

```python
%%time
arr = np.arange(size)
arr = arr * arr
```

```python
np.mean(arr)
```

```python
np.var(arr)
```

```python
np.std(arr)
```

```python
np.median(arr)
```

```python
np.percentile(arr, 50)
```

```python
np.percentile(arr, 10)
```

```python
np.percentile(arr, 90)
```

```python
iqr = np.percentile(arr, 75) - np.percentile(arr, 25)
```

```python
print(iqr)
```

```python
%%time
quartiles = np.percentile(arr, [25, 75])
iqr = quartiles[1] - quartiles[0]
```
